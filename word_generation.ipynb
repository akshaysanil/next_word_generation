```python
# import dependencies
import pandas as pd
import nltk
nltk.download('gutenberg')
from nltk.corpus import gutenberg
```

```python
# data preprocessing
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split
```

```python
with open('shakespere_hamlet.txt', 'r') as f:
    text = f.read()
text

tokenizer = Tokenizer()
tokenizer.fit_on_texts([text])
total_words = len(tokenizer.word_index) + 1
print(total_words)
```

```python
print(text)
```

```python
tokenizer.word_index
```

```python
## create input sequence
input_sequences = []

for line in text.split('\n'):
    token_list = tokenizer.texts_to_sequences([line])[0]
    # print(token_list)

    for i in range(1,len(token_list)):
        n_gram_sequence = token_list[:i+1]
        input_sequences.append(n_gram_sequence)
       
```

```python
text_1 = 'All the rage back to home'
token_list = tokenizer.texts_to_sequences([text_1])
token_list
```

```python
token_list
```

```python
len(input_sequences)
```

```python
input_sequences
```

```python
# find max length of sequence
max_sequence_len = max([len(x) for x in input_sequences])
max_sequence_len
```

```python
# create padded sequences at max length
input_sequences = np.array(pad_sequences(input_sequences, maxlen=max_sequence_len, padding='pre'))
input_sequences
```

```python
# np.array is not imp
inuput_sequences = pad_sequences(input_sequences, maxlen=max_sequence_len, padding='pre')
inuput_sequences
```

```python
# predictions and labels
import tensorflow as tf

X = input_sequences[:,:-1]
y = input_sequences[:,-1]
```

```python
len(y)
```

```python
total_words
```

```python
# covert categorical based on the words if repeating
y  = tf.keras.utils.to_categorical(y, num_classes=total_words)
    
```

```python
y.shape
```

```python
# split the model into train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
print('total_words : ',total_words)
print('max_sequence_len : ',max_sequence_len)
```

```python
# train lstm model
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout, Embedding
```

```python
model = Sequential()
model.add(Embedding(total_words,128,input_length = max_sequence_len - 1))
model.add(LSTM(150,return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(100))
model.add(Dense(total_words,activation='softmax'))
```

```python
# optimizer
from tensorflow.keras.optimizers import Adam
optimizer = Adam(learning_rate=0.001)   
```

```python


model.compile(loss='categorical_crossentropy',optimizer=optimizer,metrics=['accuracy'])
```

```python
model.summary()
```

```python
# callbacks
from tensorflow.keras.callbacks import EarlyStopping

early_stopping = EarlyStopping(monitor='val_loss',patience=10,restore_best_weights=True)
```

```python
# training
history = model.fit(X_train,y_train,epochs=100,validation_data=(X_test,y_test),batch_size=32,verbose=1)
```

```python
# save the model
model.save('word_genearation_lstm.h5')
```

```python
def generate_text(model, tokenizer, text, max_sequence_len):
    token_list = tokenizer.texts_to_sequences([text])[0]
    if len(token_list) >= max_sequence_len:
        token_list = token_list[-(max_sequence_len-1):]  # Ensure the sequence length matches max_sequence_len-1
        print('token list if >= max_sequence_len : ',token_list)
    token_list = pad_sequences([token_list],maxlen=max_sequence_len - 1)
    print('token_list : ',token_list)
    predicted = model.predict(token_list)
    print('predicted : ',predicted)
    predicted_word_index = np.argmax(predicted, axis=1)
    print('predicted_word_index : ',predicted_word_index)
    # predicted_word = tokenizer.index_word[predicted_word_index]
    # print('predicted_word : ',predicted_word)
    for word, index in tokenizer.word_index.items():
        if index == predicted_word_index:
            print('word index : ',index)
            print('word : ',word)
            print(f'{text} >> {word}')
            return word
    return None

text_1 = 'I love everything except '
text_2 = 'I love coding'
text_3 = 'coding is fun'

generate_text(model, tokenizer, text_1, max_sequence_len)
generate_text(model, tokenizer, text_2, max_sequence_len)
generate_text(model, tokenizer, text_3, max_sequence_len)
```
